```python
import pandas as pd

# Read the CSV file
csv_data = pd.read_csv("NETS_Small.csv")

# Read the TXT file with specified encoding
txt_data = pd.read_csv("Moves2020_CA(WithAddressInfo).txt", delimiter="\t", encoding='latin1')

# Combine the data frames (row-wise or column-wise based on your needs)
combined_data = pd.concat([csv_data, txt_data], axis=0)  # For stacking rows
# combined_data = pd.concat([csv_data, txt_data], axis=1)  # For merging columns

# Export combined data to a new CSV file
combined_data.to_csv("Combined_Data.csv", index=False)
```
